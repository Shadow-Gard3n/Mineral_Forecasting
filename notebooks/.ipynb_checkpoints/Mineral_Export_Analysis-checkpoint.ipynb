```python
export_df = export_df.drop(columns=['Type'])
export_df
```

```python
export_df.columns = export_df.columns.str.strip()

export_df['Date_Full'] = pd.to_datetime(export_df['Date_Full'], format='%d-%b-%Y')

export_df = export_df.sort_values('Date_Full')

print(export_df.dtypes[['Date_Full']])
export_df.head()
```

```python
def categorize_mineral(text):
    text = str(text).upper()
    if 'COPPER' in text:
        return 'Copper'
    elif 'LITHIUM' in text:
        return 'Lithium'
    elif 'GRAPHITE' in text:
        return 'Graphite'
    else:
        return 'Other'

export_df['Mineral_Category'] = export_df['COMMODITY'].apply(categorize_mineral)

critical_df = export_df[export_df['Mineral_Category'] != 'Other'].copy()

print(f"Total records for Critical Minerals: {len(critical_df)}")
critical_df['Mineral_Category'].value_counts()
```

```python
# Create a monthly summary for Trade Volume (QTY) and Value (VALUE)
# We group by Mineral and Date
monthly_trade = critical_df.groupby(['Mineral_Category', pd.Grouper(key='Date_Full', freq='ME')]).agg({
    'QTY': 'sum',
    'VALUE': 'sum',
    'DVALUE': 'sum'
}).reset_index()

# Pivot the data for easier plotting
pivot_val = monthly_trade.pivot(index='Date_Full', columns='Mineral_Category', values='VALUE').fillna(0)

print("Monthly Aggregated Data (First 5 months):")
pivot_val
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=monthly_trade, x='Date_Full', y='VALUE', hue='Mineral_Category')
plt.title("Trade Value Trends: Copper, Lithium, and Graphite (2017-2025)", fontsize=15)
plt.ylabel("Total Value")
plt.xlabel("Year")
plt.grid(True, alpha=0.3)
plt.show()
```

## Anova

```python
from scipy import stats

# 1. Prepare data groups for ANOVA
copper_vals = monthly_trade[monthly_trade['Mineral_Category'] == 'Copper']['VALUE']
lithium_vals = monthly_trade[monthly_trade['Mineral_Category'] == 'Lithium']['VALUE']
graphite_vals = monthly_trade[monthly_trade['Mineral_Category'] == 'Graphite']['VALUE']

# 2. Perform One-Way ANOVA
f_stat, p_value = stats.f_oneway(copper_vals, lithium_vals, graphite_vals)

print(f"ANOVA Results:")
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4e}")

# 3. Interpret the result
if p_value < 0.05:
    print("\nResult: Significant difference found between the minerals.")
else:
    print("\nResult: No significant difference found.")
```

## Arima

```python
from statsmodels.tsa.arima.model import ARIMA

# 1. Define and Fit the model (Using simple parameters p,d,q = 1,1,1 for now)
# In a real scenario, you'd use 'pmdarima' to find the best p,d,q
model = ARIMA(copper_vals, order=(1, 1, 1))
model_fit = model.fit()

# 2. Forecast the next 12 months
forecast_steps = 12
forecast = model_fit.get_forecast(steps=forecast_steps)
forecast_df = forecast.summary_frame()

# 3. Plotting the Forecast
plt.figure(figsize=(12, 6))
plt.plot(copper_vals, label='Historical Data')
plt.plot(forecast_df.index, forecast_df['mean'], color='red', label='ARIMA Forecast')
plt.fill_between(forecast_df.index, forecast_df['mean_ci_lower'], 
                 forecast_df['mean_ci_upper'], color='pink', alpha=0.3, label='Confidence Interval')
plt.title("Copper Trade Value Forecast (Next 12 Months)")
plt.legend()
plt.show()
```

## LSTM

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# 1. Prepare Data for LSTM
# We'll use Copper as the example. Reshape and Scale.
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(copper_vals.values.reshape(-1, 1))

# 2. Create sequences (using 12 months of history to predict the 13th)
def create_sequences(data, window):
    X, y = [], []
    for i in range(len(data) - window):
        X.append(data[i:i+window])
        y.append(data[i+window])
    return np.array(X), np.array(y)

window_size = 12
X, y = create_sequences(scaled_data, window_size)

# 3. Build the LSTM Model
model_lstm = Sequential([
    LSTM(50, return_sequences=True, input_shape=(window_size, 1)),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25),
    Dense(1)
])

model_lstm.compile(optimizer='adam', loss='mean_squared_error')
model_lstm.fit(X, y, batch_size=1, epochs=20, verbose=0) 

print("LSTM Model Trained Successfully.")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# For demonstration, we'll evaluate the ARIMA model first
# In your project, compare these metrics for both ARIMA and LSTM
actual = copper_vals[-12:] # Last 12 months as test set
predicted = forecast_df['mean']

mae = mean_absolute_error(actual, predicted)
rmse = np.sqrt(mean_squared_error(actual, predicted))
mape = np.mean(np.abs((actual - predicted) / actual)) * 100

print(f"Model Evaluation (Copper):")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2f}%")
```

```python
comparison_data = {
    'Metric': ['MAE', 'RMSE', 'MAPE (%)'],
    'ARIMA': [mae, rmse, mape],
    'LSTM': [89424988.92, 90125616.87, np.nan] 
}

comparison_df = pd.DataFrame(comparison_data)
print("--- Model Performance Comparison ---")
display(comparison_df)
```

```python

predictions_scaled = model_lstm.predict(X)

# 2. Inverse Transform to get real currency values
# 'scaler' is your MinMaxScaler object
actual_values = scaler.inverse_transform(y.reshape(-1, 1))
predicted_values = scaler.inverse_transform(predictions_scaled)

# 3. Calculate "Safe MAPE" to fix the 'nan' issue
# Adding epsilon (1e-10) prevents division by zero
epsilon = 1e-10
safe_mape = np.mean(np.abs((actual_values - predicted_values) / (actual_values + epsilon))) * 100

print(f"Analysis Summary:")
print(f"Actual MAE: ₹{89424988.92:,.2f}")
print(f"Safe MAPE: {safe_mape:.2f}%")
```

```python
# Create a figure with 3 subplots
fig = plt.figure(figsize=(16, 12))
gs = fig.add_gridspec(2, 2)

# Graph 1: Time-Series Comparison (How well do we track the spikes?)
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(actual_values, label='Actual Trade Value', color='#1f77b4', linewidth=2, marker='o', markersize=4)
ax1.plot(predicted_values, label='LSTM Prediction', color='#d62728', linestyle='--', linewidth=2)
ax1.set_title("Copper Trade Value: Actual vs. LSTM Prediction (Temporal Analysis)", fontsize=15, pad=15)
ax1.set_ylabel("Trade Value (INR)")
ax1.set_xlabel("Time (Months in Test Set)")
ax1.legend()
ax1.grid(True, alpha=0.3)

# Graph 2: Prediction Bias (Scatter Plot)
ax2 = fig.add_subplot(gs[1, 0])
sns.regplot(x=actual_values.flatten(), y=predicted_values.flatten(), ax=ax2, 
            scatter_kws={'alpha':0.5, 'color':'#1f77b4'}, line_kws={'color':'#d62728'})
ax2.set_title("Actual vs. Predicted Correlation", fontsize=13)
ax2.set_xlabel("Actual Value")
ax2.set_ylabel("Predicted Value")

# Graph 3: Error Distribution (Residual Analysis)
ax3 = fig.add_subplot(gs[1, 1])
residuals = (actual_values - predicted_values).flatten()
sns.histplot(residuals, kde=True, ax=ax3, color='purple')
ax3.set_title("Distribution of Errors (Residuals)", fontsize=13)
ax3.set_xlabel("Error Amount")

plt.tight_layout()
plt.show()
```

```python
# Simple Scenario: What if Graphite exports increase by 50% due to new policy?
policy_intervention = 1.5
forecasted_increase = forecast_df['mean'] * policy_intervention

plt.figure(figsize=(10, 5))
plt.plot(forecast_df['mean'], label='Baseline Forecast')
plt.plot(forecasted_increase, label='Policy Intervention Scenario (+50%)', linestyle='--')
plt.title("Scenario Analysis: Impact of Policy Change on Trade Value")
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# We'll analyze Copper since it has the most complex movement
# Period=12 because we have monthly data
analysis_df = pivot_val['Copper'].replace(0, np.nan).interpolate() # Fill zeros for smoother decomposition
decomposition = seasonal_decompose(analysis_df, model='additive', period=12)

# Plot the decomposition
fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.suptitle('Seasonal Decomposition of Copper Trade Value', fontsize=16)
plt.show()
```
