```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

export_df = pd.read_csv('./../dataset/Minerals_Exports_Final_Cleaned_2017_2025.csv')
import_df = pd.read_csv('./../dataset/Minerals_Imports_Final_Cleaned_2017_2025.csv')

df = pd.concat([export_df, import_df], ignore_index=True)

df['Date_Full'] = pd.to_datetime(df['Date_Full'], format='%d-%b-%Y', errors='coerce')
df['HS_CODE'] = df['HS_CODE'].astype(str)
```

# EXIM Data Analysis

## Advanced Statistical Analysis (ANOVA)

```python
# Categorize Minerals into Strategic Groups 
def get_strategic_category(hs):
    battery = ['28369100', '25041010', '26050010', '28252000'] # Lithium, Graphite, Cobalt
    base_metals = ['26030000', '26040000', '75021000', '26090000'] # Copper, Nickel, Tin
    high_tech = ['81129200', '28461010', '26159020', '28053000'] # REE, Niobium, Gallium
    
    if hs in battery: return 'Battery Minerals'
    if hs in base_metals: return 'Base Metals'
    if hs in high_tech: return 'High-Tech/Rare Earths'
    return 'Other Strategic'

df['Strategic_Category'] = df['HS_CODE'].apply(get_strategic_category)
```

```python
from scipy.stats import levene
import_only = df[df['Type'] == 'Import'].copy()

# Check if variances are equal across categories
stat, p = levene(*[import_only[import_only['Strategic_Category'] == cat]['VALUE'] 
                   for cat in import_only['Strategic_Category'].unique() if len(import_only[import_only['Strategic_Category'] == cat]) > 0])

print(f"Levene's Test P-Value: {p:.4f}")
if p < 0.05:
    print("Warning: Variances are not equal. Consider using Welch's ANOVA (pingouin.welch_anova) for more accuracy.")
```

```python
import scipy.stats as stats

# 1. Prepare groups
groups = [import_only[import_only['Strategic_Category'] == cat]['VALUE'] 
          for cat in import_only['Strategic_Category'].unique() 
          if len(import_only[import_only['Strategic_Category'] == cat]) > 0]

# 2. Kruskal-Wallis Test (The robust alternative to ANOVA)
# This test is superior here because it handles unequal variances perfectly.
h_stat, p_val = stats.kruskal(*groups)

print(f"--- Kruskal-Wallis (Robust Non-Parametric ANOVA) Results ---")
print(f"H-Statistic: {h_stat:.2f}")
print(f"P-Value: {p_val:.4e}") # Using scientific notation as P is very small

if p_val < 0.05:
    print("\nCONCLUSION: Statistically significant differences found across mineral clusters.")
    print("STRATEGIC INSIGHT: High-priority clusters like 'Battery Minerals' show distinct trade profiles ")
    print("compared to 'Base Metals', requiring specialized procurement policies.")
```

## Dependency

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def calculate_hhi_logic(sub_df):
    """Calculates Herfindahl-Hirschman Index for a filtered dataframe."""
    if sub_df.empty or sub_df['VALUE'].sum() == 0:
        return 0
    total = sub_df['VALUE'].sum()
    shares = (sub_df.groupby('CTY')['VALUE'].sum() / total) * 100
    return (shares ** 2).sum()

def calculate_shannon_logic(sub_df):
    """Calculates Shannon Diversity Index (H') for a filtered dataframe."""
    if sub_df.empty or sub_df['VALUE'].sum() == 0:
        return 0
    total = sub_df['VALUE'].sum()
    shares = sub_df.groupby('CTY')['VALUE'].sum() / total
    shares = shares[shares > 0]
    return - (shares * np.log(shares)).sum()

def calculate_strategic_dependency_extended(master_df, target_minerals_map, production_data):
    final_reports = {}

    for name, hs_code in target_minerals_map.items():
        # 1. Filter Data
        mineral_df = master_df[master_df['HS_CODE'] == str(hs_code)].copy()
        
        # 2. Financial Year Conversion
        def to_fy(date):
            if pd.isnull(date): return None
            return f"{date.year}-{str(date.year+1)[-2:]}" if date.month >= 4 else f"{date.year-1}-{str(date.year)[-2:]}"
        
        mineral_df['FY'] = mineral_df['Date_Full'].apply(to_fy)
        valid_fys = list(production_data.get(name, {}).keys())
        mineral_df = mineral_df[mineral_df['FY'].isin(valid_fys)]
        
        # 3. Aggregate Trade Quantity (Tonnes)
        trade_summary = mineral_df.groupby(['FY', 'Type'])['QTY'].sum().unstack(fill_value=0) / 1000
        for col in ['Import', 'Export']:
            if col not in trade_summary.columns: trade_summary[col] = 0
                
        # 4. NEW: Calculate Concentration Indices (HHI & Shannon)
        # We process this year by year for the specific mineral
        indices_records = []
        for fy in valid_fys:
            fy_data = mineral_df[mineral_df['FY'] == fy]
            
            # Split into Import/Export for HHI/Shannon
            imp_sub = fy_data[fy_data['Type'] == 'Import']
            exp_sub = fy_data[fy_data['Type'] == 'Export']
            
            indices_records.append({
                'FY': fy,
                'Import_HHI': calculate_hhi_logic(imp_sub),
                'Export_HHI': calculate_hhi_logic(exp_sub),
                'Import_Shannon': calculate_shannon_logic(imp_sub),
                'Export_Shannon': calculate_shannon_logic(exp_sub)
            })
        
        indices_df = pd.DataFrame(indices_records).set_index('FY')

        # 5. Integrate Production Data
        prod_df = pd.DataFrame.from_dict(production_data.get(name, {}), orient='index', columns=['Production'])
        
        # 6. Merge All Metrics
        report = trade_summary.merge(prod_df, left_index=True, right_index=True, how='outer')
        report = report.merge(indices_df, left_index=True, right_index=True, how='outer').fillna(0)
        
        # 7. Final Calculations
        report['Net_Imports'] = report['Import'] - report['Export']
        report['Total_Consumption'] = report['Production'] + report['Net_Imports']
        report['NIR_Percent'] = (report['Net_Imports'] / report['Total_Consumption'].replace(0, np.nan)) * 100
        
        # Status Labeling
        report['Status'] = report['NIR_Percent'].apply(
            lambda x: "High Dependency" if x > 50 else ("Moderate" if x > 20 else "Self-Reliant/Exporter")
        )
        
        final_reports[name] = report
        
    return final_reports

# --- Execution ---
target_map = {'Copper': '26030000', 'Lithium': '28369100', 'Graphite': '25041010'}
production_stats = {
    'Copper': {'2021-22': 115313, '2022-23': 112745, '2023-24': 125230, '2024-25': 105012},
    'Lithium': {'2021-22': 0, '2022-23': 0, '2023-24': 0, '2024-25': 0}, 
    'Graphite': {'2021-22': 62888, '2022-23': 94789, '2023-24': 169080, '2024-25': 85329}
}

# Assuming 'df' is your master dataframe
results = calculate_strategic_dependency_extended(df, target_map, production_stats)


# --- Updated Execution with Export Indices ---

# Columns to display including both Import and Export risk metrics
display_columns = [
    'Production', 
    'Import', 
    'Export', 
    'NIR_Percent', 
    'Import_HHI', 
    'Export_HHI',      # Added
    'Import_Shannon', 
    'Export_Shannon',  # Added
    'Status'
]

for mineral, table in results.items():
    print(f"\n--- {mineral} Strategic Dependency & Risk Report ---")
    
    # Create a copy for styling to avoid SettingWithCopy warnings
    styled_table = table[display_columns].copy()
    
    # 1. Format Percentages
    styled_table['NIR_Percent'] = styled_table['NIR_Percent'].map("{:.2f}%".format)
    
    # 2. Format Shannon Indices (Diversity) - Round to 2 decimal places
    styled_table['Import_Shannon'] = styled_table['Import_Shannon'].round(2)
    styled_table['Export_Shannon'] = styled_table['Export_Shannon'].round(2)
    
    # 3. Format HHI (Concentration) - Convert to Integers
    styled_table['Import_HHI'] = styled_table['Import_HHI'].astype(int)
    styled_table['Export_HHI'] = styled_table['Export_HHI'].astype(int)
    
    # 4. Format Tonnes (Optional: add commas for readability)
    styled_table['Production'] = styled_table['Production'].map("{:,.0f}".format)
    styled_table['Import'] = styled_table['Import'].map("{:,.0f}".format)
    styled_table['Export'] = styled_table['Export'].map("{:,.0f}".format)
    
    display(styled_table)
    
# display_columns = [
#     'Production', 
#     'Import', 
#     'Export', 
#     'NIR_Percent', 
#     'Import_HHI', 
#     'Import_Shannon', 
#     'Status'
# ]

# for mineral, table in results.items():
#     print(f"\n--- {mineral} Strategic Dependency & Risk Report ---")
    
#     # Optional: Rounding for cleaner display
#     styled_table = table[display_columns].copy()
#     styled_table['NIR_Percent'] = styled_table['NIR_Percent'].map("{:.2f}%".format)
#     styled_table['Import_Shannon'] = styled_table['Import_Shannon'].round(2)
#     styled_table['Import_HHI'] = styled_table['Import_HHI'].astype(int)
    
#     display(styled_table)

# # for mineral, table in results.items():
# #     print(f"\n--- {mineral} Strategic Dependency & Risk Report ---")
# #     display(table[['Production', 'NIR_Percent', 'Import_HHI', 'Import_Shannon', 'Status']])
```

## Critical Mineral Mapping

```python
# ==========================================
# FINAL CRITICAL MINERAL MAPPING FRAMEWORK
# ==========================================

# 1. Base Aggregation: Trade Volumes & Values for all minerals
mapping_df = df.groupby('COMMODITY').agg({
    'VALUE': 'sum',
    'QTY': 'sum'
}).reset_index()

# 2. Identify Key Trading Partner (Major Supplier) for each
# (Requirement: "Identify key trading partners")
major_supplier = df[df['Type'] == 'Import'].groupby('COMMODITY')['CTY'].agg(
    lambda x: x.value_counts().index[0] if not x.empty else "N/A"
).reset_index().rename(columns={'CTY': 'Major_Supplier'})

mapping_df = mapping_df.merge(major_supplier, on='COMMODITY', how='left')

# 3. Calculate Trade-based Dependency (for all 30 minerals)
trade_sum = df.groupby(['COMMODITY', 'Type'])['VALUE'].sum().unstack(fill_value=0)
mapping_df = mapping_df.merge(trade_sum[['Import', 'Export']], on='COMMODITY', how='left')
mapping_df['Trade_Dep_Ratio'] = (mapping_df['Import'] / (mapping_df['Import'] + mapping_df['Export'])).fillna(0) * 100

# 4. Integrate the "Old" Dependency Ratio (NIR %) for Copper, Lithium, Graphite
# Match the keys in your 'results' dict to the names in your 'COMMODITY' column
name_mapping = {
    'Copper': 'COPPER ORES AND CONCENTRATES',
    'Lithium': 'LITHIUM CARBONATES',
    'Graphite': 'GRAPHITE CRYSTALLINE (FLAKES/POWDER)' 
}

mapping_df['Final_Dependency_Score'] = mapping_df['Trade_Dep_Ratio'] # Default to Trade Ratio

for min_key, df_name in name_mapping.items():
    if min_key in results:
        # Get the latest NIR % from your previous calculation
        latest_nir = results[min_key]['NIR_Percent'].iloc[-1]
        mapping_df.loc[mapping_df['COMMODITY'] == df_name, 'Final_Dependency_Score'] = latest_nir

# 5. Assign Strategic Status Labels (Strategic Value 15%)
def assign_status(nir):
    if nir >= 95: return "CRITICAL (100% Import)"
    if nir > 60: return "HIGH RISK"
    if nir > 30: return "STRATEGIC"
    return "STABLE / RESILIENT"

mapping_df['Strategic_Status'] = mapping_df['Final_Dependency_Score'].apply(assign_status)

# Final Presentation
print("--- COMPREHENSIVE NATIONAL MINERAL MAPPING ---")
final_table = mapping_df[['COMMODITY', 'Major_Supplier', 'Final_Dependency_Score', 'Strategic_Status']].sort_values('Final_Dependency_Score', ascending=False)
display(final_table)

# Export this for your report
final_table.to_csv('National_Mineral_Mapping_Framework.csv', index=False)
```

## Temporal and Seasonal Trend Analysis

```python
import statsmodels.api as sm

# 1. Prepare Monthly Time-Series Data for both flows
ts_df = df.copy()
ts_df['YearMonth'] = ts_df['Date_Full'].dt.to_period('M')
ts_monthly = ts_df.groupby(['YearMonth', 'Type'])['VALUE'].sum().unstack(fill_value=0)
ts_monthly.index = ts_monthly.index.to_timestamp()

# 2. Define a function for Dual Decomposition Visualization
def plot_trade_decomposition(series, flow_type, color):
    # period=12 for monthly data (yearly cycle)
    res = sm.tsa.seasonal_decompose(series, model='additive', period=12)
    
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(14, 12), sharex=True)
    
    # Observed
    ax1.plot(series, label=f'Original {flow_type} Value', color=color, linewidth=2)
    ax1.set_title(f'Temporal Analysis: {flow_type} Dynamics (2017-Present)', fontsize=14, fontweight='bold')
    
    # Trend (Structural Direction)
    ax2.plot(res.trend, label='Structural Trend', color='red', linestyle='--')
    
    # Seasonal (Cyclical Patterns)
    ax3.plot(res.seasonal, label='Seasonal Cycle (Annual Patterns)', color='green')
    
    # Residuals (Structural Shocks)
    ax4.scatter(series.index, res.resid, label='Residuals (Policy Shocks/External Factors)', alpha=0.6, color='purple')
    ax4.axhline(0, color='black', linestyle='-')
    
    for ax in [ax1, ax2, ax3, ax4]:
        ax.legend(loc='upper left')
        ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Run decomposition for both Imports and Exports
plot_trade_decomposition(ts_monthly['Import'], 'Import', 'blue')
plot_trade_decomposition(ts_monthly['Export'], 'Export', 'darkorange')

print("[INSIGHT]: Compare the 'Residuals' of both plots. If they spike at the same time (e.g., early 2020), the shock was global. If only Exports spike (e.g., late 2023), it may indicate a domestic policy change like an export tax.")
```

## Machine Learning-Based Forecasting

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input

# Helper to calculate evaluation metrics
def calculate_metrics(actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mape = np.mean(np.abs((actual - predicted) / (actual + 1e-10))) * 100
    r2 = r2_score(actual, predicted)
    return {"MAE": mae, "RMSE": rmse, "MAPE": mape, "R2": r2}

def run_validated_strategic_forecast(series, name="Trade"):
    # 1. PRE-PROCESSING
    # Fill gaps and ensure monthly frequency
    series = series.resample('ME').sum().interpolate(method='linear')
    
    # 2. STL DECOMPOSITION (Trend + Seasonal)
    stl = STL(series, period=12, robust=True)
    res = stl.fit()
    trend = res.trend
    seasonal = res.seasonal
    
    # 3. DATA SCALING (Using RobustScaler for large Import values)
    scaler = RobustScaler()
    trend_scaled = scaler.fit_transform(trend.values.reshape(-1, 1))
    
    # Create Sequences for LSTM
    seq_len = 6
    X, y = [], []
    for i in range(len(trend_scaled) - seq_len):
        X.append(trend_scaled[i:i + seq_len])
        y.append(trend_scaled[i + seq_len])
    X, y = np.array(X), np.array(y)
    
    # --- PHASE 1: VALIDATION ---
    # Split: Use all data except the last 12 months for training
    test_size = 12 
    X_train, y_train = X[:-test_size], y[:-test_size]
    X_test = X[-test_size:]
    actual_test_raw = series.values[-test_size:]
    
    # Build and Train Model
    model = Sequential([
        Input(shape=(seq_len, 1)),
        LSTM(64, activation='tanh'),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mae')
    model.fit(X_train, y_train, epochs=100, batch_size=4, verbose=0)
    
    # Predict Trend for Test period
    trend_test_scaled = model.predict(X_test, verbose=0)
    trend_test_unscaled = scaler.inverse_transform(trend_test_scaled).flatten()
    
    # Recombine with Seasonal for Final Test Prediction
    seasonal_test = seasonal.values[-test_size:]
    test_predictions = trend_test_unscaled + seasonal_test
    
    # Calculate Metrics
    m = calculate_metrics(actual_test_raw, test_predictions)
    
    # --- PHASE 2: FUTURE FORECASTING ---
    # Update model with full data for future prediction
    model.fit(X, y, epochs=20, batch_size=4, verbose=0)
    
    steps = 36
    curr_seq = trend_scaled[-seq_len:].reshape(1, seq_len, 1)
    future_trend_scaled = []
    for _ in range(steps):
        nxt = model.predict(curr_seq, verbose=0)[0, 0]
        future_trend_scaled.append(nxt)
        curr_seq = np.append(curr_seq[:, 1:, :], [[[nxt]]], axis=1)
        
    future_trend = scaler.inverse_transform(np.array(future_trend_scaled).reshape(-1, 1)).flatten()
    future_seasonal = np.tile(seasonal.values[-12:], 3)[:steps]
    future_final = future_trend + future_seasonal
    
    # Connect future to the last actual point
    future_dates = pd.date_range(start=series.index[-1], periods=steps + 1, freq='ME')
    future_values = np.insert(future_final, 0, series.values[-1])

    # 5. VISUALIZATION
    plt.figure(figsize=(15, 8))
    
    # Plot Historical Training Data
    plt.plot(series.index[:-test_size], series.values[:-test_size], color='black', alpha=0.2, label='Historical (Train)')
    
    # Plot Actual Test Data (Validation)
    plt.plot(series.index[-test_size:], actual_test_raw, color='black', marker='o', markersize=4, label='Actual (Test Period)')
    
    # Plot Model's Test Prediction (The Validation Line)
    plt.plot(series.index[-test_size:], test_predictions, color='green', linestyle='--', lw=2.5, label='Model Validation Line')
    
    # Plot Future Forecasts
    plt.plot(future_dates, future_values, color='red', lw=2, label='Medium-Term Forecast (3-Year)')
    plt.plot(future_dates[:7], future_values[:7], color='blue', lw=4, label='Short-Term Forecast (6-Month)')
    
    # Shading
    plt.axvspan(series.index[-test_size], series.index[-1], color='green', alpha=0.05, label='Validation Zone')
    
    # Title with Metrics
    plt.title(f"Strategic {name} Forecast with Visual Validation\nMAPE: {m['MAPE']:.2f}% | R2: {m['R2']:.2f}", fontsize=16)
    plt.ylabel('Value')
    plt.ticklabel_format(style='plain', axis='y') # Fix for large Import numbers
    plt.grid(True, linestyle='--', alpha=0.4)
    plt.legend(loc='upper left', ncol=2)
    
    plt.tight_layout()
    plt.show()

    print(f"--- {name} Metrics ---")
    print(m)

# Execute for both
run_validated_strategic_forecast(ts_monthly['Import'], "Import")
run_validated_strategic_forecast(ts_monthly['Export'], "Export")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input

# --- 1. CONFIGURATION: HS_CODE PREFIXES ---
# Mapping the first few digits of HS_CODEs
mineral_mapping = {
    "COPPER": "74",      # HS Codes starting with 74
    "LITHIUM": "2836",   # Commonly Lithium Carbonate
    "GRAPHITE": "2504"   # Natural Graphite
}

def run_hscode_strategic_forecast(series, name):
    if series.empty or series.sum() == 0:
        print(f"Skipping {name}: No trade data found for this HS_CODE.")
        return

    # Pre-processing
    series = series.resample('ME').sum().interpolate(method='linear')
    
    # STL Decomposition
    stl = STL(series, period=12, robust=True)
    res = stl.fit()
    trend, seasonal = res.trend, res.seasonal
    
    # LSTM on Trend
    scaler = RobustScaler()
    trend_scaled = scaler.fit_transform(trend.values.reshape(-1, 1))
    
    seq_len = 6
    X, y = [], []
    for i in range(len(trend_scaled) - seq_len):
        X.append(trend_scaled[i:i + seq_len])
        y.append(trend_scaled[i + seq_len])
    X, y = np.array(X), np.array(y)
    
    # Validation Split
    test_size = 12
    X_train, y_train = X[:-test_size], y[:-test_size]
    X_test, actual_test = X[-test_size:], series.values[-test_size:]
    
    # Build Model
    model = Sequential([Input(shape=(seq_len, 1)), LSTM(32), Dense(1)])
    model.compile(optimizer='adam', loss='mae')
    model.fit(X_train, y_train, epochs=100, batch_size=4, verbose=0)
    
    # Validation Prediction
    trend_test = scaler.inverse_transform(model.predict(X_test, verbose=0)).flatten()
    val_preds = trend_test + seasonal.values[-test_size:]
    r2 = r2_score(actual_test, val_preds)

    # Future Forecast (36m)
    steps = 36
    curr_seq = trend_scaled[-seq_len:].reshape(1, seq_len, 1)
    f_trend_scaled = []
    for _ in range(steps):
        nxt = model.predict(curr_seq, verbose=0)[0, 0]
        f_trend_scaled.append(nxt)
        curr_seq = np.append(curr_seq[:, 1:, :], [[[nxt]]], axis=1)
        
    f_trend = scaler.inverse_transform(np.array(f_trend_scaled).reshape(-1, 1)).flatten()
    f_seasonal = np.tile(seasonal.values[-12:], 3)[:steps]
    future_final = f_trend + f_seasonal
    
    # Anchoring for plot
    f_dates = pd.date_range(start=series.index[-1], periods=steps + 1, freq='ME')
    f_values = np.insert(future_final, 0, series.values[-1])

    # --- PLOTTING ---
    plt.figure(figsize=(14, 6))
    plt.plot(series.index, series.values, color='black', alpha=0.15, label='History')
    plt.plot(series.index[-test_size:], actual_test, color='black', marker='o', markersize=3, label='Actual (Test Period)')
    plt.plot(series.index[-test_size:], val_preds, color='green', linestyle='--', label='Model Validation Line')
    
    plt.plot(f_dates, f_values, color='red', lw=2, label='Medium-Term (3-Year)')
    plt.plot(f_dates[:7], f_values[:7], color='blue', lw=4, label='Short-Term (6-Month)')
    
    plt.title(f"{name} Strategy: HS_CODE Prefix Forecast (R2: {r2:.2f})", fontsize=14)
    plt.ticklabel_format(style='plain', axis='y')
    plt.grid(True, alpha=0.3)
    plt.legend(loc='upper left', ncol=2)
    plt.show()

# --- EXECUTION ---
for mineral, prefix in mineral_mapping.items():
    # Filter by HS_CODE prefix
    # (Assuming HS_CODE is a string column in your 'df')
    el_df = df[df['HS_CODE'].str.startswith(prefix, na=False)]
    
    # Aggregate
    ts = el_df.groupby([pd.Grouper(key='Date_Full', freq='ME'), 'Type'])['VALUE'].sum().unstack(fill_value=0)
    
    for trade_type in ['Import', 'Export']:
        if trade_type in ts.columns:
            run_hscode_strategic_forecast(ts[trade_type], f"{mineral} {trade_type}")
```
