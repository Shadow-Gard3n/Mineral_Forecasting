```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error
from scipy import stats
import logging

# Reproducibility
np.random.seed(42)
torch.manual_seed(42)
sns.set_theme(style="whitegrid")

logging.basicConfig(level=logging.INFO, format='%(message)s')
logger = logging.getLogger()
```

```python
def load_and_clean_data(export_path, import_path):
    try:
        export_df = pd.read_csv(export_path)
        import_df = pd.read_csv(import_path)
    except FileNotFoundError as e:
        print(f"Error: {e}")
        return None

    df = pd.concat([export_df, import_df], ignore_index=True)
    initial_count = len(df)

    # Convert Date
    df['Date_Full'] = pd.to_datetime(df['Date_Full'], format='%d-%b-%Y', errors='coerce')
    
    df['YEAR'] = df['Date_Full'].dt.year

    df['HS_CODE'] = df['HS_CODE'].astype(str).str.strip().str.zfill(8)
    
    for col in ['VALUE', 'QTY', 'YEAR']:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    df.dropna(subset=['Date_Full', 'VALUE', 'HS_CODE'], inplace=True)
    print(f"Loaded {len(df)} rows (Dropped {initial_count - len(df)} invalid rows).")
    
    return df

df = load_and_clean_data('./../dataset/Minerals_Exports_Final_Cleaned_2017_2025.csv', 
                         './../dataset/Minerals_Imports_Final_Cleaned_2017_2025.csv')
```

## Advanced Statistical Analysis (ANOVA)

```python
STRATEGIC_MAP = {
    'Battery Minerals': ['28369100', '25041010', '26050010', '28252000'],
    'Base Metals': ['26030000', '26040000', '75021000', '26090000'],
    'High-Tech/REE': ['81129200', '28461010', '26159020', '28053000']
}

hs_lookup = {code: cat for cat, codes in STRATEGIC_MAP.items() for code in codes}

df['Strategic_Category'] = df['HS_CODE'].map(hs_lookup).fillna('Other Strategic')

validation_summary = df.groupby('Strategic_Category')['HS_CODE'].nunique()
print("--- Mineral Categorization Validation ---")
print(validation_summary)
```

```python
from scipy.stats import levene
import_only = df[df['Type'] == 'Import'].copy()

# Check if variances are equal across categories
stat, p = levene(*[import_only[import_only['Strategic_Category'] == cat]['VALUE'] 
                   for cat in import_only['Strategic_Category'].unique() if len(import_only[import_only['Strategic_Category'] == cat]) > 0])

print(f"Levene's Test P-Value: {p:.4f}")
if p < 0.05:
    print("Warning: Variances are not equal. Consider using Welch's ANOVA (pingouin.welch_anova) for more accuracy.")
```

```python
from pingouin import welch_anova

print("--- Normality Check (Shapiro-Wilk) ---")
for cat in df['Strategic_Category'].unique():
    sample = df[df['Strategic_Category'] == cat]['VALUE'].sample(min(100, len(df)))
    stat, p = stats.shapiro(sample)
    print(f"{cat}: p-value={p:.4f} ({'Normal' if p > 0.05 else 'Non-Normal'})")

aov = welch_anova(dv='VALUE', between='Strategic_Category', data=df[df['Type'] == 'Import'])
print("\n--- Welch's ANOVA Results ---")
aov
```

```python
import scipy.stats as stats

groups = [import_only[import_only['Strategic_Category'] == cat]['VALUE'] 
          for cat in import_only['Strategic_Category'].unique() 
          if len(import_only[import_only['Strategic_Category'] == cat]) > 0]


h_stat, p_val = stats.kruskal(*groups)

print(f"--- Kruskal-Wallis (Robust Non-Parametric ANOVA) Results ---")
print(f"H-Statistic: {h_stat:.2f}")
print(f"P-Value: {p_val:.4e}") # Using scientific notation as P is very small

if p_val < 0.05:
    print("\nCONCLUSION: Statistically significant differences found across mineral clusters.")
    print("STRATEGIC INSIGHT: High-priority clusters like 'Battery Minerals' show distinct trade profiles ")
    print("compared to 'Base Metals', requiring specialized procurement policies.")
```

## Dependency

```python
import pandas as pd
import numpy as np

def calculate_risk_indices(data):
    """Calculates HHI and Shannon Diversity with safety checks."""
    if data.empty:
        return np.nan, np.nan, "No Data"
        
    partner_shares = data.groupby('CTY')['VALUE'].sum()
    total_val = partner_shares.sum()
    
    if total_val <= 0:
        return np.nan, np.nan, "No Data"
    
    shares = partner_shares / total_val
    hhi = (shares**2).sum() * 10000
    shannon = - (shares * np.log(shares + 1e-9)).sum()
    
    if hhi > 2500: status = "High Risk"
    elif hhi > 1500: status = "Medium Risk"
    else: status = "Low Risk"
    
    return hhi, shannon, status

def calculate_strategic_dependency_extended(master_df, target_minerals_map, production_data):
    final_reports = {}

    for name, hs_codes in target_minerals_map.items():
        if isinstance(hs_codes, str): hs_codes = [hs_codes]
        
        # 1. Filter and Clean
        mineral_df = master_df[master_df['HS_CODE'].isin(hs_codes)].copy()
        
        # Financial Year Conversion
        def to_fy(date):
            if pd.isnull(date): return None
            return f"{date.year}-{str(date.year+1)[-2:]}" if date.month >= 4 else f"{date.year-1}-{str(date.year)[-2:]}"
        
        mineral_df['FY'] = mineral_df['Date_Full'].apply(to_fy)
        valid_fys = list(production_data.get(name, {}).keys())
        mineral_df = mineral_df[mineral_df['FY'].isin(valid_fys)]
        
        # 2. Aggregate Trade Quantity (Tonnes)
        trade_summary = mineral_df.groupby(['FY', 'Type'])['QTY'].sum().unstack(fill_value=0) / 1000
        for col in ['Import', 'Export']:
            if col not in trade_summary.columns: trade_summary[col] = 0.0
                
        # 3. Calculate Concentration Indices
        indices_records = []
        for fy in valid_fys:
            fy_data = mineral_df[mineral_df['FY'] == fy]
            imp_hhi, imp_sha, imp_stat = calculate_risk_indices(fy_data[fy_data['Type'] == 'Import'])
            exp_hhi, exp_sha, _ = calculate_risk_indices(fy_data[fy_data['Type'] == 'Export'])
            
            indices_records.append({
                'FY': fy,
                'Import_HHI': imp_hhi, 'Export_HHI': exp_hhi,
                'Import_Shannon': imp_sha, 'Export_Shannon': exp_sha,
                'Risk_Status': imp_stat
            })
        
        indices_df = pd.DataFrame(indices_records).set_index('FY')

        # 4. Merge Production & Compute NIR
        prod_df = pd.DataFrame.from_dict(production_data.get(name, {}), orient='index', columns=['Production'])
        report = trade_summary.merge(prod_df, left_index=True, right_index=True, how='outer')
        report = report.merge(indices_df, left_index=True, right_index=True, how='outer')
        
        report['Net_Imports'] = report['Import'] - report['Export']
        report['Total_Consumption'] = report['Production'] + report['Net_Imports']
        
        # NIR Formula: (Net Imports / Apparent Consumption)
        report['NIR_Percent'] = (report['Net_Imports'] / report['Total_Consumption'].replace(0, np.nan)) * 100
        
        report['Dependency_Status'] = report['NIR_Percent'].apply(
            lambda x: "High Dependency" if x > 50 else ("Moderate" if x > 20 else "Self-Reliant/Exporter")
        )
        
        final_reports[name] = report
        
    return final_reports

# --- Execution ---

production_stats = {
    'Copper': {'2021-22': 115313, '2022-23': 112745, '2023-24': 125230, '2024-25': 105012},
    'Lithium': {'2021-22': 0, '2022-23': 0, '2023-24': 0, '2024-25': 0}, 
    'Graphite': {'2021-22': 62888, '2022-23': 94789, '2023-24': 169080, '2024-25': 85329}
}

target_map = {
    'Copper': ['26030000', '74031100'],
    'Lithium': ['25309099', '28369100', '28252000'],
    'Graphite': ['25041010', '25041020', '38011000']
}

results = calculate_strategic_dependency_extended(df, target_map, production_stats)

for mineral, table in results.items():
    print(f"\n--- {mineral} Strategic Dependency & Risk Report ---")
    
    cols = ['Production', 'Import', 'Export', 'NIR_Percent', 'Import_HHI', 'Export_HHI', 
            'Import_Shannon', 'Export_Shannon', 'Risk_Status', 'Dependency_Status']
    
    out = table[cols].copy()
    
    out['NIR_Percent'] = out['NIR_Percent'].map(lambda x: f"{x:.1f}%" if pd.notnull(x) else "N/A")
    out['Import_HHI'] = out['Import_HHI'].map(lambda x: f"{int(x)}" if pd.notnull(x) else "N/A")
    out['Export_HHI'] = out['Export_HHI'].map(lambda x: f"{int(x)}" if pd.notnull(x) else "N/A")
    out['Import_Shannon'] = out['Import_Shannon'].map(lambda x: f"{x:.2f}" if pd.notnull(x) else "N/A")
    out['Export_Shannon'] = out['Export_Shannon'].map(lambda x: f"{x:.2f}" if pd.notnull(x) else "N/A")
    
    display(out)
```

## Critical Mineral Mapping

```python
# FINAL CRITICAL MINERAL MAPPING FRAMEWORK


# 1. Base Aggregation: Trade Volumes & Values for all minerals
mapping_df = df.groupby('COMMODITY').agg({
    'VALUE': 'sum',
    'QTY': 'sum'
}).reset_index()

# 2. Identify Key Trading Partner (Major Supplier) for each
major_supplier = df[df['Type'] == 'Import'].groupby('COMMODITY')['CTY'].agg(
    lambda x: x.value_counts().index[0] if not x.empty else "N/A"
).reset_index().rename(columns={'CTY': 'Major_Supplier'})

mapping_df = mapping_df.merge(major_supplier, on='COMMODITY', how='left')

# 3. Calculate Trade-based Dependency (for all 30 minerals)
trade_sum = df.groupby(['COMMODITY', 'Type'])['VALUE'].sum().unstack(fill_value=0)
mapping_df = mapping_df.merge(trade_sum[['Import', 'Export']], on='COMMODITY', how='left')
mapping_df['Trade_Dep_Ratio'] = (mapping_df['Import'] / (mapping_df['Import'] + mapping_df['Export'])).fillna(0) * 100

# 4. Integrate the "Old" Dependency Ratio (NIR %) for Copper, Lithium, Graphite
name_mapping = {
    'Copper': 'COPPER ORES AND CONCENTRATES',
    'Lithium': 'LITHIUM CARBONATES',
    'Graphite': 'GRAPHITE CRYSTALLINE (FLAKES/POWDER)' 
}

mapping_df['Final_Dependency_Score'] = mapping_df['Trade_Dep_Ratio'] # Default to Trade Ratio

for min_key, df_name in name_mapping.items():
    if min_key in results:
        latest_nir = results[min_key]['NIR_Percent'].iloc[-1]
        mapping_df.loc[mapping_df['COMMODITY'] == df_name, 'Final_Dependency_Score'] = latest_nir

# 5. Assign Strategic Status Labels (Strategic Value 15%)
def assign_status(nir):
    if nir >= 95: return "CRITICAL (100% Import)"
    if nir > 60: return "HIGH RISK"
    if nir > 30: return "STRATEGIC"
    return "STABLE / RESILIENT"

mapping_df['Strategic_Status'] = mapping_df['Final_Dependency_Score'].apply(assign_status)

# Final Presentation
print("--- COMPREHENSIVE NATIONAL MINERAL MAPPING ---")
final_table = mapping_df[['COMMODITY', 'Major_Supplier', 'Final_Dependency_Score', 'Strategic_Status']].sort_values('Final_Dependency_Score', ascending=False)
display(final_table)

# Export this for your report
# final_table.to_csv('National_Mineral_Mapping_Framework.csv', index=False)
```

## Temporal and Seasonal Trend Analysis

```python
import statsmodels.api as sm

# 1. Prepare Monthly Time-Series Data for both flows
ts_df = df.copy()
ts_df['YearMonth'] = ts_df['Date_Full'].dt.to_period('M')
ts_monthly = ts_df.groupby(['YearMonth', 'Type'])['VALUE'].sum().unstack(fill_value=0)
ts_monthly.index = ts_monthly.index.to_timestamp()

# 2. Define a function for Dual Decomposition Visualization
def plot_trade_decomposition(series, flow_type, color):
    # period=12 for monthly data (yearly cycle)
    res = sm.tsa.seasonal_decompose(series, model='additive', period=12)
    
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(14, 12), sharex=True)
    
    # Observed
    ax1.plot(series, label=f'Original {flow_type} Value', color=color, linewidth=2)
    ax1.set_title(f'Temporal Analysis: {flow_type} Dynamics (2017-Present)', fontsize=14, fontweight='bold')
    
    # Trend (Structural Direction)
    ax2.plot(res.trend, label='Structural Trend', color='red', linestyle='--')
    
    # Seasonal (Cyclical Patterns)
    ax3.plot(res.seasonal, label='Seasonal Cycle (Annual Patterns)', color='green')
    
    # Residuals (Structural Shocks)
    ax4.scatter(series.index, res.resid, label='Residuals (Policy Shocks/External Factors)', alpha=0.6, color='purple')
    ax4.axhline(0, color='black', linestyle='-')
    
    for ax in [ax1, ax2, ax3, ax4]:
        ax.legend(loc='upper left')
        ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Run decomposition for both Imports and Exports
plot_trade_decomposition(ts_monthly['Import'], 'Import', 'blue')
plot_trade_decomposition(ts_monthly['Export'], 'Export', 'darkorange')

print("[INSIGHT]: Compare the 'Residuals' of both plots. If they spike at the same time (e.g., early 2020), the shock was global. If only Exports spike (e.g., late 2023), it may indicate a domestic policy change like an export tax.")
```

## Machine Learning-Based Forecasting

#### Traditional method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, r2_score

def run_basic_arima_comparison(series, name):
    # 1. Resample to Monthly (ME) for consistency and stability
    series_m = series.resample('ME').sum().interpolate()
    
    if len(series_m) < 15:
        print(f"SKIPPING {name}: Not enough data points.")
        return

    # 2. Temporal Split (80% Train, 20% Test)
    train_size = int(len(series_m) * 0.8)
    train, test = series_m.iloc[:train_size], series_m.iloc[train_size:]
    
    # 3. Fit ARIMA Model
    # order=(1,1,1) is a robust starting point for trade data
    try:
        model = ARIMA(train, order=(1, 1, 1))
        model_fit = model.fit()
        
        # 4. Predict on the Test Set
        predictions = model_fit.get_forecast(steps=len(test)).predicted_mean
        predictions.index = test.index
        
        # 5. Calculate Metrics
        mape = mean_absolute_percentage_error(test, predictions)
        rmse = np.sqrt(mean_squared_error(test, predictions))
        r2 = r2_score(test, predictions)
        
        print(f"\n--- ARIMA Metrics: {name} ---")
        print(f"MAPE: {mape:.2%}")
        print(f"RMSE: {rmse:,.0f}")
        print(f"R2 Score: {r2:.2f}")

        # 6. Visualization: Actual vs. Predicted
        plt.figure(figsize=(12, 5))
        plt.plot(train.index, train, label='Training Data', color='black', alpha=0.3)
        plt.plot(test.index, test, label='Actual Test Data', color='black', marker='o', markersize=4)
        plt.plot(test.index, predictions, label='ARIMA Prediction', color='blue', linestyle='--', lw=2)
        
        plt.title(f"ARIMA Comparison: {name}\nMAPE: {mape:.2%} | R2: {r2:.2f}")
        plt.ylabel('Value')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()
        
    except Exception as e:
        print(f"Error modeling {name}: {e}")

# --- EXECUTION LOOP ---
target_map = {
    'Copper': ['26030000', '74031100'],
    'Lithium': ['25309099', '28369100', '28252000'],
    'Graphite': ['25041010', '25041020', '38011000']
}

for mineral_name, hs_codes in target_map.items():
    mask = df['HS_CODE'].astype(str).isin([str(c) for c in hs_codes])
    mineral_df = df[mask].copy()
    
    for trade_type in ['Import', 'Export']:
        subset = mineral_df[mineral_df['Type'] == trade_type].copy()
        
        if subset.empty:
            print(f"No data found for {mineral_name} {trade_type}")
            continue
        
        # --- Aggregation ---
        # If Copper Ore and Copper Cathode both have data for Jan 2024, 
        # we sum them together to get the total 'Copper' value.
        subset = subset.groupby('Date_Full')['VALUE'].sum().reset_index()
        subset.set_index('Date_Full', inplace=True)
        run_basic_arima_comparison(subset['VALUE'], f"{mineral_name} {trade_type}")

        
```

#### Modern method

```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_percentage_error, r2_score
from statsmodels.tsa.seasonal import STL

# Standard LSTM Architecture
class TrendLSTM(nn.Module):
    def __init__(self, input_size=1, hidden_size=32, output_size=1):
        super(TrendLSTM, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out, _ = self.lstm(x)
        return self.fc(out[:, -1, :])

def run_strategic_forecast_with_plots(series, name, units="USD"):
    # --- PRE-PROCESSING & LOG TRANSFORM ---
    series_m = series.resample('ME').sum()
    series_log = np.log1p(series_m)
    
    if len(series_m) < 24:
        print(f"SKIPPING {name}: Need at least 24 months for STL Decomposition.")
        return

    # --- DECOMPOSITION (In Log Space) ---
    stl = STL(series_log, period=12, robust=True)
    result = stl.fit()
    trend, seasonal = result.trend, result.seasonal

    # --- DATA PREP ---
    train_size = int(len(trend) * 0.8)
    train_trend = trend.iloc[:train_size].values.reshape(-1, 1)
    test_trend = trend.iloc[train_size:].values.reshape(-1, 1)
    
    scaler = MinMaxScaler()
    train_trend_scaled = scaler.fit_transform(train_trend)
    test_trend_scaled = scaler.transform(test_trend)
    
    window = 6 
    
    def create_xy(data, win):
        x, y = [], []
        for i in range(len(data)-win):
            x.append(data[i:i+win]); y.append(data[i+win])
        return (torch.FloatTensor(np.array(x)), torch.FloatTensor(np.array(y))) if x else (None, None)

    X_train, y_train = create_xy(train_trend_scaled, window)
    if X_train is None: return

    # --- TRAINING ---
    model = TrendLSTM(input_size=1, hidden_size=32)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.005)
    criterion = nn.HuberLoss() 
    
    model.train()
    for _ in range(300):
        optimizer.zero_grad()
        loss = criterion(model(X_train), y_train)
        loss.backward(); optimizer.step()
        
    # --- VALIDATION & FUTURE FORECASTING ---
    model.eval()
    with torch.no_grad():
        val_input = np.vstack([train_trend_scaled[-window:], test_trend_scaled])
        X_test, _ = create_xy(val_input, window)
        val_trend_preds = scaler.inverse_transform(model(X_test).numpy()).flatten()
        
        future_steps = 36
        curr_seq = torch.FloatTensor(test_trend_scaled[-window:].reshape(1, window, 1))
        future_trend_scaled = []
        for _ in range(future_steps):
            pred = model(curr_seq)
            future_trend_scaled.append(pred.item())
            curr_seq = torch.cat((curr_seq[:, 1:, :], pred.unsqueeze(1)), dim=1)
        future_trend = scaler.inverse_transform(np.array(future_trend_scaled).reshape(-1, 1)).flatten()

    # --- RECOMBINE & METRICS ---
    val_final_preds = np.expm1(val_trend_preds + seasonal.iloc[train_size:].values)
    future_final = np.expm1(future_trend + np.tile(seasonal.iloc[-12:].values, int(future_steps/12)))
    actual_test = series_m.iloc[train_size:].values

    # Safety metrics for divide-by-zero
    sum_actual = np.sum(actual_test)
    wape = np.sum(np.abs(actual_test - val_final_preds)) / sum_actual if sum_actual > 0 else 0
    bias = np.sum(val_final_preds - actual_test) / sum_actual if sum_actual > 0 else 0
    
    naive_mae = np.mean(np.abs(np.diff(series_m.iloc[:train_size].values)))
    mase = np.mean(np.abs(actual_test - val_final_preds)) / naive_mae if naive_mae > 0 else 0
    rmse = np.sqrt(np.mean((actual_test - val_final_preds)**2))
    r2 = r2_score(actual_test, val_final_preds)

    # --- VISUALIZATION ---
    test_dates = series_m.index[train_size:]
    future_dates = pd.date_range(start=test_dates[-1] + pd.offsets.MonthEnd(1), periods=future_steps, freq='ME')

    plt.figure(figsize=(15, 7))
    plt.plot(series_m.index[:train_size], series_m.values[:train_size], color='black', alpha=0.15, label='Historical')
    plt.plot(test_dates, actual_test, color='black', marker='o', markersize=4, label='Actual (Test)')
    plt.plot(test_dates, val_final_preds, color='green', linestyle='--', lw=2, label='Recombined Validation')
    plt.plot(future_dates, future_final, color='red', lw=2, label='Future Forecast (3-Year)')
    
    plt.axvspan(test_dates[0], test_dates[-1], color='green', alpha=0.05, label='Validation Zone')
    title_str = (f"{name} Forecast ({units})\n"
                 f"WAPE: {wape:.2%} | MASE: {mase:.2f} | Bias: {bias:.2%}\n"
                 f"RMSE: {rmse:,.0f} | R2: {r2:.2f}")
    plt.title(title_str, fontsize=13, loc='left', fontweight='bold')
    plt.legend(loc='upper left', ncol=2)
    plt.grid(True, alpha=0.2)
    plt.tight_layout()
    plt.show()

# Run the groups
target_map = {
    'Copper': ['26030000', '74031100'],
    'Lithium': ['25309099', '28369100', '28252000'],
    'Graphite': ['25041010', '25041020', '38011000']
}

for mineral_name, hs_codes in target_map.items():
    mask = df['HS_CODE'].astype(str).isin([str(c) for c in hs_codes])
    mineral_df = df[mask].copy()
    for trade_type in ['Import', 'Export']:
        subset = mineral_df[mineral_df['Type'] == trade_type].copy()
        if subset.empty: continue
        subset = subset.groupby('Date_Full')['VALUE'].sum().reset_index()
        subset.set_index('Date_Full', inplace=True)
        run_strategic_forecast_with_plots(subset['VALUE'], f"{mineral_name} {trade_type}")
```
