```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from pmdarima import auto_arima
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# 2. Pipeline Cleaning Function
def clean_pipeline(df, trade_type):
    df['Type'] = trade_type
    df['Date_Full'] = pd.to_datetime(df['Date_Full'], format='%d-%b-%Y', errors='coerce')
    df['HS_CODE'] = df['HS_CODE'].astype(str)
    # Calculate Unit Price (Strategic Requirement)
    df['Unit_Price'] = df['VALUE'] / df['QTY'].replace(0, np.nan)
    return df

export_df = clean_pipeline(export_df, 'Export')
import_df = clean_pipeline(import_df, 'Import')
df = pd.concat([export_df, import_df], ignore_index=True).dropna(subset=['Date_Full'])

print("Pipeline initialized. Total records:", len(df))
```

```python
# Create Dependency Matrix
dep_matrix = df.groupby(['COMMODITY', 'Type'])['VALUE'].sum().unstack(fill_value=0)
dep_matrix['Net_Dependency'] = dep_matrix['Import'] - dep_matrix['Export']
dep_matrix['Dependency_Index'] = (dep_matrix['Import'] / (dep_matrix['Import'] + dep_matrix['Export'])) * 100
dep_matrix = dep_matrix.sort_values(by='Import', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=dep_matrix['Import'].head(10), y=dep_matrix.head(10).index, palette='Reds_r')
plt.title('Top 10 Critical Minerals by Import Value (India)', fontsize=14)
plt.show()

display(dep_matrix.head(10))
```

```python
def run_mineral_forecast(hs_code, mineral_name):
    print(f"\n--- Analyzing {mineral_name} ({hs_code}) ---")
    
    # 1. Isolate Imports & Reindex for continuity
    ts = df[(df['HS_CODE'] == hs_code) & (df['Type'] == 'Import')].groupby('Date_Full')['VALUE'].sum()
    full_range = pd.date_range(start=ts.index.min(), end=ts.index.max(), freq='MS')
    ts = ts.reindex(full_range, fill_value=0)
    
    # 2. Stationarity Check
    p_val = adfuller(ts.diff().dropna())[1]
    print(f"Stationarity p-value (after diff): {p_val:.4f}")

    # 3. ARIMA Forecast (12 Months)
    model_arima = auto_arima(ts, seasonal=True, m=12, suppress_warnings=True)
    arima_forecast = model_arima.predict(n_periods=12)
    
    # 4. LSTM Prep & Forecast
    scaler = MinMaxScaler()
    scaled = scaler.fit_transform(ts.values.reshape(-1, 1))
    
    # Simple window-based forecast logic
    X = np.array([scaled[i:i+12] for i in range(len(scaled)-12)])
    y = scaled[12:]
    
    model_lstm = Sequential([
        LSTM(50, activation='relu', input_shape=(12, 1)),
        Dense(1)
    ])
    model_lstm.compile(optimizer='adam', loss='mse')
    model_lstm.fit(X, y, epochs=50, verbose=0)
    
    # Predict future
    last_win = scaled[-12:].reshape(1, 12, 1)
    lstm_f = []
    for _ in range(12):
        p = model_lstm.predict(last_win, verbose=0)
        lstm_f.append(p[0,0])
        last_win = np.append(last_win[:,1:,:], [p], axis=1)
    
    lstm_forecast = scaler.inverse_transform(np.array(lstm_f).reshape(-1, 1))
    
    # 5. Strategic Visualization
    plt.figure(figsize=(14, 5))
    plt.plot(ts.index, ts, label='Historical', color='blue')
    f_dates = pd.date_range(start=ts.index[-1]+pd.DateOffset(months=1), periods=12, freq='MS')
    plt.plot(f_dates, arima_forecast, label='ARIMA Forecast', linestyle='--', color='red')
    plt.plot(f_dates, lstm_forecast, label='LSTM Forecast', linestyle='--', color='orange')
    plt.title(f'12-Month Import Forecast: {mineral_name}')
    plt.legend()
    plt.show()

# Run the pipeline for Top Minerals
run_mineral_forecast('28369100', 'Lithium Carbonate')
run_mineral_forecast('26030000', 'Copper Ores')
```
