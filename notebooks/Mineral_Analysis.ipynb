```python
export_df['HS_CODE'] = export_df['HS_CODE'].astype(str)
import_df['HS_CODE'] = import_df['HS_CODE'].astype(str)

export_df['Date_Full'] = pd.to_datetime(export_df['Date_Full'])
import_df['Date_Full'] = pd.to_datetime(import_df['Date_Full'])

df = pd.concat([export_df, import_df], ignore_index=True)

df.head()
```

```python
monthly_trade = df.groupby(['Date_Full', 'Type']).agg({
    'VALUE': 'sum',
    'QTY': 'sum'
}).reset_index()

trade_pivot = monthly_trade.pivot(index='Date_Full', columns='Type', values='VALUE').fillna(0)

# 5. Calculate the Net Trade Balance
# Formula: Export Value - Import Value
trade_pivot['Net_Balance'] = trade_pivot['Export'] - trade_pivot['Import']
```

```python
sns.lineplot(data=trade_pivot, x=trade_pivot.index, y='Export', label='Total Exports', color='green', linewidth=2.5)
sns.lineplot(data=trade_pivot, x=trade_pivot.index, y='Import', label='Total Imports', color='red', linewidth=2.5)

# Highlighting the Trade Deficit/Surplus area
plt.fill_between(trade_pivot.index, trade_pivot['Export'], trade_pivot['Import'], 
                 where=(trade_pivot['Export'] < trade_pivot['Import']), 
                 color='red', alpha=0.15, label='Trade Deficit')
plt.fill_between(trade_pivot.index, trade_pivot['Export'], trade_pivot['Import'], 
                 where=(trade_pivot['Export'] >= trade_pivot['Import']), 
                 color='green', alpha=0.15, label='Trade Surplus')

# Formatting the Chart
plt.title('India\'s Critical Mineral Trade: Exports vs Imports (2017-2025)', fontsize=16, fontweight='bold')
plt.ylabel('Trade Value', fontsize=12)
plt.xlabel('Timeline', fontsize=12)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)

# Note: The gap before 2021 exists because Import data starts from 2021
plt.tight_layout()
plt.show()

# Display the numerical summary for the last 5 months
print("Recent Monthly Trade Summary:")
print(trade_pivot.tail())
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose

# 1. Prepare Time Series for a specific Critical Mineral (e.g., Lithium Carbonates)
# We use the pivoted data but filtered for a specific commodity
# lithium_ts = df[df['HS_CODE'] == '28369100'].groupby('Date_Full')['VALUE'].sum()
# Filter specifically for 'Import' to analyze India's dependency
lithium_ts = df[(df['HS_CODE'] == '28369100') & (df['Type'] == 'Import')].groupby('Date_Full')['VALUE'].sum()
# 2. Visualize the raw trend
plt.figure(figsize=(12, 5))
lithium_ts.plot(title='Monthly Import Value Trend: Lithium Carbonates', color='blue')
plt.ylabel('Value')
plt.show()

# 3. Perform ADF Test (Stationarity Check)
# Required by PDF: "Handle time series data with proper preprocessing (stationarity checks)"
print("--- Augmented Dickey-Fuller Test ---")
result = adfuller(lithium_ts.dropna())
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
if result[1] <= 0.05:
    print("Result: Data is Stationary (Ready for Modeling)")
else:
    print("Result: Data is NOT Stationary (Needs Differencing before ARIMA)")

# 4. Seasonal Decomposition
# Helps identify "Seasonal patterns and cyclical trends" required by PDF
decomposition = seasonal_decompose(lithium_ts, model='additive', period=12)
decomposition.plot()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller

# 1. Select Commodity (Ensure it matches your HS_CODE type - we use 'str' as you did)
target_code = '28369100' # Lithium Carbonates

# 2. Extract and fill missing months to ensure a continuous timeline
# This prevents errors in seasonal_decompose
raw_series = df[(df['HS_CODE'] == target_code) & (df['Type'] == 'Import')].groupby('Date_Full')['VALUE'].sum()

# Create a full range of months from the first to the last available date
full_range = pd.date_range(start=raw_series.index.min(), end=raw_series.index.max(), freq='MS')
lithium_ts = raw_series.reindex(full_range, fill_value=0)

# 3. Seasonal Decomposition
# Note: Requires at least 24 months for 12-month seasonality
if len(lithium_ts) >= 24:
    decomposition = seasonal_decompose(lithium_ts, model='additive', period=12)
    fig = decomposition.plot()
    fig.set_size_inches(12, 8)
    plt.suptitle(f'Decomposition of {target_code} Imports', fontsize=16)
    plt.show()
else:
    print("Error: Not enough data points (need at least 24 months).")

# 4. Handling Non-Stationarity (Differencing)
# Since your p-value was > 0.05, we must "Difference" the data
print("\n--- Fixing Non-Stationarity ---")
lithium_diff = lithium_ts.diff().dropna()

result = adfuller(lithium_diff)
print(f'New ADF Statistic (After 1st Difference): {result[0]}')
print(f'New p-value: {result[1]}')

if result[1] <= 0.05:
    print("SUCCESS: Data is now stationary and ready for ARIMA modeling.")
else:
    print("ACTION: Data still not stationary. Try a 2nd difference or Log Transformation.")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler

# 1. Data Preparation
# Scale data between 0 and 1 for the Neural Network
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(lithium_ts.values.reshape(-1, 1))

# Create sequences (e.g., use 12 months of data to predict the 13th)
def create_sequences(data, window):
    X, y = [], []
    for i in range(len(data) - window):
        X.append(data[i:i+window])
        y.append(data[i+window])
    return np.array(X), np.array(y)

WINDOW_SIZE = 12
X, y = create_sequences(scaled_data, WINDOW_SIZE)

# Split into train/test (match your ARIMA 6-month test split)
split = len(X) - 6
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# 2. Build the LSTM Model
model = Sequential([
    LSTM(50, activation='relu', input_shape=(WINDOW_SIZE, 1), return_sequences=True),
    Dropout(0.2),
    LSTM(50, activation='relu'),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# 3. Train the Model
print("Training LSTM Model...")
history = model.fit(X_train, y_train, epochs=100, batch_size=4, validation_data=(X_test, y_test), verbose=0)

# 4. Predictions & Inverse Scaling
lstm_predictions = model.predict(X_test)
lstm_predictions = scaler.inverse_transform(lstm_predictions)
y_test_actual = scaler.inverse_transform(y_test)

# 5. Compare Results
lstm_rmse = np.sqrt(mean_squared_error(y_test_actual, lstm_predictions))
print(f"\n--- Model Comparison ---")
print(f"ARIMA RMSE: {136756228.62}") # From your previous result
print(f"LSTM RMSE: {lstm_rmse:.2f}")

# Visualization of Comparison
plt.figure(figsize=(12, 6))
plt.plot(test.index, y_test_actual, label='Actual Values', color='black', linewidth=2)
plt.plot(test.index, lstm_predictions, label='LSTM Forecast', color='orange', linestyle='--')
plt.plot(test.index, test_predictions, label='ARIMA Forecast', color='red', linestyle='--')
plt.title('Comparison: ARIMA vs LSTM (Test Set Performance)')
plt.legend()
plt.show()
```
